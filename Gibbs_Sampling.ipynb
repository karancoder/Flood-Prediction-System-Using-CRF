```
from scipy.interpolate import RegularGridInterpolator

def regrid(data, out_x, out_y):
    m = max(data.shape[0], data.shape[1])
    y = np.linspace(0, 1.0/m, data.shape[0])
    x = np.linspace(0, 1.0/m, data.shape[1])
    interpolating_function = RegularGridInterpolator((y, x), data)

    yv, xv = np.meshgrid(np.linspace(0, 1.0/m, out_y), np.linspace(0, 1.0/m, out_x))

    return interpolating_function((xv, yv))
```

```
class CRF:

  y_potential_array = np.array([[1, 1], [1, 1]])
  x_potential_array = np.array([[1, 1], [1, 1]])


  def get_random(self):								# function to get a random number
    RAND_MAX = 10000
    val1 = float(np.random.rand() % RAND_MAX)
    val2 = float(RAND_MAX)
    return val1/val2
  
  def searchNearbyJunction(self, ind_i, ind_j, junction_mat, tol=2):
    if junction_mat[ind_i,ind_j] == 1:
      return True
    # for temp_i in range(ind_i - tol, ind_i + tol + 1):
    #   for temp_j in range(ind_j - tol, ind_j + tol + 1):
    #     if junctions_mat[min(max(0, temp_i), junctions_mat.shape[0])][min(max(0, temp_j), junctions_mat.shape[1])]:
    #       return True
    return False

  def yPotential(self, x, y, beta, offset):
    potential = np.zeros((beta.shape[0]))
    # print((2*x + y))
    potential[int((4*offset) + (2*x) + y)] = beta[int((4*offset) + (2*x) + y)]
    return potential

  # change this
  def alt_function(self, alt1, alt2):
    if float(alt2 - alt1) > 0.0:
      return 0
    elif float(alt2 - alt1) == 0.0:
      return 1
    else:
      return 2

  def getYPotential(self, y, ind_i, ind_j, beta, alt_mat):
    potential = np.zeros((beta.shape[0]))
    if ind_i > 0:
      potential += self.yPotential(y[ind_i][ind_j], y[ind_i - 1][ind_j], beta, self.alt_function(alt_mat[ind_i][ind_j], alt_mat[ind_i - 1][ind_j]))
      # potential += self.yPotential(y[ind_i][ind_j], y[ind_i - 1][ind_j], beta)
    if ind_i < y.shape[0] - 1:
      potential += self.yPotential(y[ind_i][ind_j], y[ind_i + 1][ind_j], beta, self.alt_function(alt_mat[ind_i][ind_j], alt_mat[ind_i + 1][ind_j]))
      # potential += self.yPotential(y[ind_i][ind_j], y[ind_i + 1][ind_j], beta)
    if ind_j > 0:
      potential += self.yPotential(y[ind_i][ind_j], y[ind_i][ind_j - 1], beta, self.alt_function(alt_mat[ind_i][ind_j], alt_mat[ind_i][ind_j - 1]))
      # potential += self.yPotential(y[ind_i][ind_j], y[ind_i][ind_j - 1], beta)
    if ind_j < y.shape[1] - 1:
      potential += self.yPotential(y[ind_i][ind_j], y[ind_i][ind_j + 1], beta, self.alt_function(alt_mat[ind_i][ind_j], alt_mat[ind_i][ind_j + 1]))
      # potential += self.yPotential(y[ind_i][ind_j], y[ind_i][ind_j + 1], beta)
    return potential
  
  def getXPotential(self, x, y, ind_i, ind_j, alpha, junction_mat, prec_mat):
    potential = np.zeros((alpha.shape[0]))
    a = x[ind_i][ind_j]
    b = y[ind_i][ind_j]
    c = self.searchNearbyJunction(ind_i, ind_j, junction_mat)
    d = 0 # np.argmax(prec_mat[:,ind_i,ind_j])
    for l in range(prec_mat.shape[0]):
      if prec_mat[l][ind_i][ind_j] == 1:
        d = l
        break
    potential[int(2 * a + b)] = alpha[int(2 * a + b)]
    potential[4 + int(2 * c + b)] = alpha[4 + int(2 * c + b)]
    potential[8 + int(2 * d + b)] = alpha[8 + int(2 * d + b)]
    return potential
  
  def getEnergy(self, x, y, ind_i, ind_j, alpha, beta, junction_mat, alt_mat, prec_mat):
    energy = np.zeros((alpha.shape[0] + beta.shape[0]))
    temp_energy = self.getYPotential(y, ind_i, ind_j, beta, alt_mat)
    # print(ind_i, ind_j, temp_energy)
    energy[alpha.shape[0]:] += temp_energy
    temp_energy = self.getXPotential(x, y, ind_i, ind_j, alpha, junction_mat, prec_mat)
    energy[:alpha.shape[0]] += temp_energy
    return energy
  
  def getEnergyTotal(self, x, y, alpha, beta, junction_mat, alt_mat, prec_mat):
    total_energy = np.zeros((alpha.shape[0] + beta.shape[0]))
    # print(total_energy)
    for i in range(prec_mat.shape[0]):
      total_energy[8 + (2 * i)] += alpha[(2 * i) + 8] * np.count_nonzero((prec_mat[i][:y.shape[0],:y.shape[1]] == 1) & (y == 0))
      total_energy[8 + (2 * i) + 1] += alpha[(2 * i) + 1 + 8] * np.count_nonzero((prec_mat[i][:y.shape[0],:y.shape[1]] == 1) & (y == 1))
    for i in range(4):
      a = int(i / 2)
      b = i % 2
      total_energy[i] += alpha[i] * np.count_nonzero((x == a) & (y == b))
      total_energy[4 + i] += alpha[i + 4] * np.count_nonzero((junction_mat[:y.shape[0],:y.shape[1]] == a) & (y == b))
      total_energy[alpha.shape[0] + i] += beta[i] * (
          np.count_nonzero((alt_mat[:,:-1] > alt_mat[:,1:]) & (y[:,:-1] == a) & (y[:,1:] == b)) + \
          np.count_nonzero((alt_mat[:-1,:] > alt_mat[1:,:]) & (y[:-1,:] == a) & (y[1:,:] == b)) + \
          np.count_nonzero((alt_mat[:,1:] > alt_mat[:,:-1]) & (y[:,1:] == a) & (y[:,:-1] == b)) + \
          np.count_nonzero((alt_mat[1:,:] > alt_mat[:-1,:]) & (y[1:,:] == a) & (y[:-1,:] == b))
      )
      total_energy[alpha.shape[0] + 4 + i] += beta[4 + i] * (
          np.count_nonzero((alt_mat[:,:-1] == alt_mat[:,1:]) & (y[:,:-1] == a) & (y[:,1:] == b)) + \
          np.count_nonzero((alt_mat[:-1,:] == alt_mat[1:,:]) & (y[:-1,:] == a) & (y[1:,:] == b)) + \
          np.count_nonzero((alt_mat[:,1:] == alt_mat[:,:-1]) & (y[:,1:] == a) & (y[:,:-1] == b)) + \
          np.count_nonzero((alt_mat[1:,:] == alt_mat[:-1,:]) & (y[1:,:] == a) & (y[:-1,:] == b))
      )
      total_energy[alpha.shape[0] + 8 + i] += beta[8 + i] * (
          np.count_nonzero((alt_mat[:,:-1] < alt_mat[:,1:]) & (y[:,:-1] == a) & (y[:,1:] == b)) + \
          np.count_nonzero((alt_mat[:-1,:] < alt_mat[1:,:]) & (y[:-1,:] == a) & (y[1:,:] == b)) + \
          np.count_nonzero((alt_mat[:,1:] < alt_mat[:,:-1]) & (y[:,1:] == a) & (y[:,:-1] == b)) + \
          np.count_nonzero((alt_mat[1:,:] < alt_mat[:-1,:]) & (y[1:,:] == a) & (y[:-1,:] == b))
      )
    # for i in range(y.shape[0]):
    #   for j in range(y.shape[1]):
    #     temp_energy = self.getEnergy(x, y, i, j, alpha, beta)
    #     total_energy += temp_energy
    # print(total_energy)
    return total_energy

  # change this function
  def updateSample(self, x, y, alpha, beta, junction_mat, alt_mat, prec_mat):
    # total_energy = np.zeros((alpha.shape[0] + beta.shape[0]))
    # print(total_energy)
    i_ind_range = list(range(y.shape[0]))
    random.shuffle(i_ind_range)
    j_ind_range = list(range(y.shape[1]))
    random.shuffle(j_ind_range)
    for i in i_ind_range:
      for j in j_ind_range:
        temp_energy_current = self.getEnergy(x, y, i, j, alpha, beta, junction_mat, alt_mat, prec_mat)
        cell_energy_current = np.sum(temp_energy_current)
        y[i][j] += 1
        y[i][j] %= 2
        temp_energy_prime = self.getEnergy(x, y, i, j, alpha, beta, junction_mat, alt_mat, prec_mat)
        cell_energy_prime = np.sum(temp_energy_prime)

        prime_prob = math.exp(- cell_energy_prime) / (math.exp(- cell_energy_prime) + math.exp(- cell_energy_current))

        if np.random.rand() > prime_prob: #+ (0.00 * cell_energy_current):
          y[i][j] += 1
          y[i][j] %= 2
        #   total_energy += temp_energy_current
        # else:
        #   total_energy += temp_energy_prime
    return y
  
  def get_tot_details(self, sample,y,val):								# Function used to find the confusion_matrix, accuracy_score and classification report of the sample data compared with the is_flooded data(stored in y)
    s = sample.shape
    jval = int(s[0]/2)
    y1 = y.flatten()
    s1 = sample.flatten()
    if (val == 0):
      y = (y[:jval]).flatten()
      sample = (sample[:jval]).flatten()
    elif (val == -1):
      y = y.flatten()
      sample = sample.flatten()
    else :
      y = (y[jval:]).flatten()
      sample = (sample[jval:]).flatten()
    mat = confusion_matrix(y,sample) 
    acc = accuracy_score(y, sample)
    clr = classification_report(y,sample)
    return [mat,acc,clr]

  def print_accs(self, t1):												# Function used to print the confusion_matrix, accuracy_score and classification_report into the terminal in a more structured way
    print('Confusion Matrix :')
    print(t1[0]) 
    print('Accuracy Score :' + str(t1[1])) 
    print('Report : ')
    print(t1[2])

  def train_model(self, x, y, junction_mat, alt_mat, prec_mat, max_iterations,gibbs_sampling_rate,learning_rate,alpha=None,beta=None,samples=None,test=False, x_test=None, y_test=None, junction_mat_test=None, alt_mat_test=None, prec_mat_test=None):				# Function used to train the model on the compressed data set

    y_s = y.shape
    if samples is None:
      samples = []
    if alpha is None:
      alpha = np.array([self.get_random() for i in range(8 + (2 * prec_mat.shape[0]))])
    if beta is None:
      beta = np.array([self.get_random() for i in range(12)])
    print("Alpha: ", alpha, " Beta: ", beta)

    start_time = time.perf_counter()
    ini_model_val = self.getEnergyTotal(x, y, np.ones((8 + (2 * prec_mat.shape[0]))), np.ones(12), junction_mat, alt_mat, prec_mat)
    print("In Potential: ", ini_model_val)
    end_time = time.perf_counter()
    print((end_time - start_time))

    prev_best_acc = 0
    # sample = np.random.randint(2,size = (y_s[0],y_s[1]))

    for its in range(max_iterations):
      start_time = time.perf_counter()
      print(its,end = " : ")
      total_sample = np.zeros((y_s[0],y_s[1]))
      if its % (max_iterations / 10) == 0:
        # sample = np.random.randint(2,size = (y_s[0],y_s[1]))
        sample = np.zeros(y_s)
      # plt.imshow(sample)
      sum_val = np.zeros((alpha.shape[0] + beta.shape[0]))
      for it in range(gibbs_sampling_rate):
        start_time = time.perf_counter()
        # print(it,end = ":")
        sample = self.updateSample(x, sample, alpha, beta, junction_mat, alt_mat, prec_mat)
        # print("su",end = ":")
        total_sample += sample
        sum_val += self.getEnergyTotal(x, sample, np.ones((8 + (2 * prec_mat.shape[0]))), np.ones(12), junction_mat, alt_mat, prec_mat)
        # print("ea",end = ",")
        end_time = time.perf_counter()
        print((end_time - start_time))
      print(" ")
      sample = (total_sample / gibbs_sampling_rate).round()
      end_time = time.perf_counter()
      print((end_time - start_time))
      start_time = time.perf_counter()
      # plt.imshow(sample)
      t1 = [tr_mat,tr_acc,tr_clr] = self.get_tot_details(sample,y,-1)
      new_sum_val = sum_val / gibbs_sampling_rate
      print("In Potential: ", ini_model_val, "Sample Potential: ", new_sum_val)
      self.print_accs(t1)
      alpha -= learning_rate * (ini_model_val[:alpha.shape[0]] - new_sum_val[:alpha.shape[0]]) / (y_s[0] * y_s[1])
      beta -= learning_rate * (ini_model_val[alpha.shape[0]:] - new_sum_val[alpha.shape[0]:]) / (y_s[0] * y_s[1])
      print("Alpha: ", alpha, " Beta: ", beta)
      print("parameters updated")
      samples.append(sample)
      end_time = time.perf_counter()
      print((end_time - start_time))
      
      # sample = np.random.randint(2,size = (y_s[0],y_s[1]))
      # for it in range(gibbs_sampling_rate):
      #   # print(it,end = ":")
      #   sample = self.updateSample(x, sample, alpha, beta)
      
      # t2 = [te_mat,te_acc,te_clr] = self.get_tot_details(sample,y,-1)
      # self.print_accs(t2)
      # if (te_acc > prev_best_acc):
      #   # write_asc_data("prediction_vals.asc",sample,y_x1,y_y1,y_cell_size)
      #   prev_best_acc = te_acc	
    if test is True:
      sample = np.random.randint(2,size = y_test.shape)
      sample = np.zeros(y_test.shape)
      total_sample = np.zeros(y_test.shape)
      for it in range(gibbs_sampling_rate * max(1, int(max_iterations / 10.0))):
        sample = self.updateSample(x_test, sample, alpha, beta, junction_mat_test, alt_mat_test, prec_mat_test)
        total_sample += sample
      sample = (total_sample / gibbs_sampling_rate).round()
      # print(ini_model_val, self.getEnergyTotal(x, sample, np.ones((4)), np.ones(4)))
      t1 = [tr_mat,tr_acc,tr_clr] = self.get_tot_details(sample,y_test,-1)
      # t2 = [te_mat,te_acc,te_clr] = self.get_tot_details(sample,y,1)
      self.print_accs(t1)
      # plt.imshow(sample)
      # plt.colorbar()
      samples.append(sample)
      # self.print_accs(t2)
      # if (te_acc > prev_best_acc):
        # write_asc_data("prediction_vals.asc",sample,y_x1,y_y1,y_cell_size)
        # prev_best_acc = te_acc
    return samples, alpha, beta
  
  def test_model(self,gibbs_sampling_rate,alpha=None,beta=None,samples=None,x_test=None, y_test=None, junction_mat_test=None, alt_mat_test=None, prec_mat_test=None):				# Function used to test the model
    sample = np.zeros(y_test.shape)
    if samples is None or samples == []:
      samples = []
    else:
      sample = samples[-1]
    
    print("Alpha: ", alpha, " Beta: ", beta)
    start_time = time.perf_counter()
    # sample = np.random.randint(2,size = y_test.shape)
    total_sample = np.zeros(y_test.shape)
    for it in range(gibbs_sampling_rate):
      sample = self.updateSample(x_test, sample, alpha, beta, junction_mat_test, alt_mat_test, prec_mat_test)
      total_sample += sample
      samples.append(sample)
    sample = (total_sample / gibbs_sampling_rate).round()
    t1 = [tr_mat,tr_acc,tr_clr] = self.get_tot_details(sample,y_test,-1)
    self.print_accs(t1)
    # plt.imshow(sample)
    # plt.colorbar()
    # samples.append(sample)
    end_time = time.perf_counter()
    print((end_time - start_time))
    return samples, alpha, beta
```

```
def manhattanDist(x1, y1, x2, y2):
  return abs(x1 - x2) + abs(y1 - y2)

def createTiltedSquare(len_of_mat, half_length, noise):
  x = np.zeros((len_of_mat, len_of_mat))
  for i in range(len_of_mat):
    for j in range(len_of_mat):
      if (i >= 2 and j >= 2 and i <= len_of_mat - 3 and j <= len_of_mat - 3):
        if(manhattanDist(i, j, len_of_mat / 2.0, len_of_mat / 2.0) <= half_length):
          x[i,j] = 1
      if np.random.rand() <= noise:
        x[i,j] = (x[i,j] + 1) % 2
  return x
x = createTiltedSquare(20,7,0.05)
plt.imshow(x)
plt.figure()
y = createTiltedSquare(20,7,0)
plt.imshow(y)
```

```
def gaussianDist(x1, y1, x2, y2):
  return np.sqrt([((x1 - x2) ** 2) + ((y1 - y2) ** 2)])

def createCircle(len_of_mat, outer_radius, inner_radius, noise):
  x = np.zeros((len_of_mat, len_of_mat))
  for i in range(len_of_mat):
    for j in range(len_of_mat):
      if (i >= 2 and j >= 2 and i <= len_of_mat - 3 and j <= len_of_mat - 3):
        if (gaussianDist(i, j, len_of_mat / 2.0, len_of_mat / 2.0) >= inner_radius) and (gaussianDist(i, j, len_of_mat / 2.0, len_of_mat / 2.0) <= outer_radius):
          x[i,j] = 1
      if np.random.rand() <= noise:
        x[i,j] = (x[i,j] + 1) % 2
  return x
x = createCircle(20,7,4,0.05)
plt.imshow(x)
plt.figure()
y = createCircle(20,7,4,0)
plt.imshow(y)
```

```
def createCross(len_of_cross, noise):
  x = np.zeros((len_of_cross, len_of_cross))
  for i in range(len_of_cross):
    for j in range(len_of_cross):
      if (i >= 2 and j >= 2 and i <= len_of_cross - 3 and j <= len_of_cross - 3):
        if (i == j or i == j - 1 or i == j + 1):
          x[i,j] = 1
        if (i + j == len_of_cross - 1 or i + j == len_of_cross - 2 or i + j == len_of_cross):
          x[i,j] = 1
      if np.random.rand() <= noise:
        x[i,j] = (x[i,j] + 1) % 2
  return x
x = createCross(20,0.05)
plt.imshow(x)
plt.figure()
y = createCross(20,0)
plt.imshow(y)
```
